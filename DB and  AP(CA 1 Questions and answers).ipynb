**a).Create a function to import the web-3Dayforecast.xml XML data file. Your function should include exception

```python
import xml.etree.ElementTree as ET

def import_weather_data(file_path):
# To parse the xml file using Element tree
    try:
        tree = ET.parse(file_path)
# To get the root element        
        root = tree.getroot()
        return root
    except FileNotFoundError:
# If the file is not found
        print(f"Error: File not found at {file_path}")
    except ET.ParseError:
# If there are any parsing errors
        print(f"Error: Unable to parse XML file at {file_path}")
    except Exception as e:
# If there are any errors
        print(f"An unexpected error occurred: {e}")

file_path = r"C:\Users\musta\Downloads\NCI DATA ANALYTICS\datasets\web-3Dayforecast.xml"
# Calling the function
weather_data = import_weather_data(file_path)

# To Check if the import was successful
if weather_data is not None:
    print("XML file imported successfully.")
    
else:
    print("Failed to import XML file.")
```

**c) Calculate the average max_temp for across all days for stations where the id value is greater than 03956. 
[Note: you can re-use the loop from part b) if you wish.]**

```python
import xml.etree.ElementTree as ET

def calculate_average_max_temp(xml_root):
    total_max_temp = 0
    count = 0

    # Iterate over each station in the XML data
    for station in xml_root.findall('.//station'):
        station_id = int(station.find('id').text)

        # Check if the station ID is greater than 03956
        if station_id > 3956:
            # Iterate over each day within the station
            for day in station.findall('day'):
                max_temp = int(day.find('max_temp').text)
                total_max_temp += max_temp
                count += 1

    # Calculate the average max_temp
    if count > 0:
        average_max_temp = total_max_temp / count
        return average_max_temp
    else:
        return None

# Example usage
file_path = r"C:\Users\musta\Downloads\NCI DATA ANALYTICS\datasets\web-3Dayforecast.xml"
weather_data = import_weather_data(file_path)

if weather_data is not None:
    # Call the function to calculate average max_temp
    average_max_temp = calculate_average_max_temp(weather_data)

    if average_max_temp is not None:
        print(f"Average Max Temp for stations with ID > 03956: {average_max_temp:.2f}")
    else:
        print("No matching stations found.")
else:
    print("Failed to import XML file.")
```

**d) Extract the following information from the XML weather forecast data and write it to a CSV file: 
• Id 
• location 
• day_num 
• wind_speed 
• wind_dir 
Your file should also contain the column names.**

```python
import os
import csv
import xml.etree.ElementTree as ET

def extract_and_write_to_csv(xml_root, csv_filename):
    # Extract the directory path from the full file path
    output_directory = os.path.dirname(csv_filename)

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Open a CSV file for writing
    with open(csv_filename, 'w', newline='') as csvfile:
        # Define the CSV writer
        csv_writer = csv.writer(csvfile)

        # Write column names to the CSV file
        csv_writer.writerow(['Id', 'Location', 'Day_Num', 'Wind_Speed', 'Wind_Dir'])

        # Iterate over each station in the XML data
        for station in xml_root.findall('.//station'):
            station_id = station.find('id').text
            location = station.find('location').text

            # Iterate over each day within the station
            for day in station.findall('day'):
                day_num = day.find('day_num').text
                wind_speed = day.find('wind_speed').text
                wind_dir = day.find('wind_dir').text

                # Write the extracted information to the CSV file
                csv_writer.writerow([station_id, location, day_num, wind_speed, wind_dir])

# Example usage
file_path =  r"C:\Users\musta\Downloads\NCI DATA ANALYTICS\datasets\web-3Dayforecast.xml"
output_csv_filename = "output/weather_data.csv"
weather_data = import_weather_data(file_path)

if weather_data is not None:
    # Call the function to extract and write to CSV
    extract_and_write_to_csv(weather_data, output_csv_filename)

    print(f"Data successfully written to {output_csv_filename}")
else:
    print("Failed to import XML file.")
```

## Question 2

**Given the following string: 
GPT don't chat
To me or yourself  
Unless it is noon 
On the Moon. 
a) Using a single regular expression, write a function to highlight all occurrences of words that consist of exactly 
two letters. Your answer should look like this:  
GPT don't chat 
{To} {me} {or} yourself  
Unless {it} {is} noon 
On the Moon.**

```python
import re

def highlight_two_letter_words(text):
    # Define the regular expression pattern to match two-letter words
    pattern = r'\b\w{2}\b'

    # Use re.sub to replace matched words with curly braces
    result = re.sub(pattern, lambda match: '{' + match.group() + '}', text)

    return result

# Example usage
input_text = """GPT don't chat 
To me or yourself  
Unless it is noon 
On the Moon."""

highlighted_text = highlight_two_letter_words(input_text)
print(highlighted_text)
```

**b) Using another single regular expression, write another function that highlights all the first three characters on 
each line. Your answer should look like this: 
{GPT} don't chat 
{To }me or yourself  
{Unl}ess it is noon 
{On }the Moon.**

```python
def highlight_first_three_characters(text):
    # Define the regular expression pattern to match the first three characters on each line
    pattern = r'^(.{3})'

    # Use re.sub to replace matched patterns with curly braces
    result = re.sub(pattern, lambda match: '{' + match.group(1) + '}', text, flags=re.MULTILINE)

    return result

# Example usage
input_text = """GPT don't chat 
To me or yourself  
Unless it is noon 
On the Moon."""

highlighted_text = highlight_first_three_characters(input_text)
print(highlighted_text)
```

```python
import re

def re_show(pat, s):
    print(re.compile(pat, re.M).sub("{\g<0>}", s.rstrip()))
    print("\n")

s = """ GPT don't chat 
To me or yourself  
Unless it is noon 
On the Moon."""

# Define your regular expressions
myRegexA = r'\b\w{2}\b'  # Regex for part a) - Highlighting two-letter words
myRegexB = r'^(.{3})'    # Regex for part b) - Highlighting the first three characters on each line

# Use the re_show function
re_show(myRegexA, s)
re_show(myRegexB, s)
```

**c) Extract all the XML data and write it to a CSV file. Your file should also contain the column names.**

```python
import csv

csv_file = "./theodata.csv"

try :
    theo_csv = open(csv_file, 'w')
    csvwriter = csv.writer(theo_csv)
    head = []
    rowcount = 0
    for subject in theo_xml.findall('subject'):
        row = []
        if rowcount == 0:
            for col in range(0,len(list(subject))) :
                head.append(subject[col].tag)
            csvwriter.writerow(head)
            rowcount = rowcount + 1
        for col in range(0,len(list(subject))) :
            row.append(subject[col].text)
        csvwriter.writerow(row)
except FileNotFoundError:
    print('Could not open the file:'.format(csv_file))
except IOError:
    print('Could not write to the file:'.format(csv_file))
finally :
    theo_csv.close()
```
